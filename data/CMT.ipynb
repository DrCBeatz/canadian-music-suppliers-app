```python
from CMT import CMT

CMT
```

```python
CMT = CMT.split(') ')
```

```python
CMT
```

```python
vendors = {}
for line in CMT:
    vendor = line.split(' (')[0]
    try:
        supplier = line.split(' (')[1]
    except:
        supplier = ''
    vendors[vendor] = supplier

vendors
```

```python
len(vendors)
```

```python
import pandas as pd

df = pd.DataFrame.from_dict(vendors, orient='index')
df = df.reset_index()
df.columns = ['Vendor', 'Supplier']
df.head()
```

```python
df.to_csv('CMT.csv', index=False)
```
